### Importing Data from CSV file 

```python
import pandas as pd # importing pandas module for data analysis tasks 
import matplotlib.pyplot as plt

x = pd.read_csv("C:\\Users\\moomo\\OneDrive\\Desktop\\fraud_data.csv") # function in pandas--read_csv("file path")

x.head(2)

```

### Plotting Line Graph 

```python
plt.plot(x.trans_time, x.amt, 'ro--')

plt.title('Transactions at different Times')
plt.xlim(-0.4,8)
plt.ylim(-1,100)
plt.xlabel('Time of Transaction')
plt.ylabel('AMT of Transaction')



plt.show()
```

### Plotting Bar Graph

```python
x.head()

plt.bar(x.category, x.amt)
plt.xlabel('Category')
plt.ylabel('Amount Spent') # shopping category has most exspenditure --makes sense, fraud transactions usually spent on shopping 
plt.title('$ Spent per Category')

# ~$392 million spent on fraud in the world 

plt.show()
```
